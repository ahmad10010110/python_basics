### EX09
Find max of Total.Population and subtract all values of Total.Population column from the maximum value using lambda function.


```python
#first of all we have to change this column name 
newds = ds.rename(columns={ds.columns[3] : "TotalPopulation"})
#finding max
mx = newds["TotalPopulation"].max()

newds.TotalPopulation.map(lambda i: i - mx)
```
